# Deployed Jupyter: Saving Your Notebooks

This notebook explains how to save work in the deployed JupyterLab environment and where to get help.

## Where files persist

- Deployed notebooks are stored on persistent storage mounted at `/workspace`.
- Files under `/workspace/notebooks` should remain available across container restarts.
- Keep your work in `/workspace/notebooks` for a predictable location.

## How to save reliably in the deployed app

1. Create or open notebooks in `/workspace/notebooks`.
2. Save frequently with **Ctrl+S** (or **File -> Save Notebook**).
3. Rename notebooks from `Untitled.ipynb` to meaningful names.
4. Refresh the browser and confirm the file is still listed.

## If you cannot open a notebook

- Refresh JupyterLab once and retry.
- Reopen the notebook from the left file browser at `/notebooks`.
- If the error persists, capture the exact URL and timestamp and contact support.

## Where to get help

For issues with saves, missing files, stale content, or node integration:

- Start from the orchestrator app and related service links.
- Contact the Labmarket platform team through your normal support channel.
- Include: notebook path, exact URL, UTC timestamp, and any error text.
